```python
fig=plt.figure(figsize=(15,8))
data1 = deliveries.groupby(['match_id','batsman'],as_index=False).batsman_runs.sum().batsman_runs  
data2 = deliveries.groupby(['match_id','batsman'],as_index=False).ball.count().ball > 20
data3  = data1[data2 == True]
data4 = deliveries.groupby(['match_id','batsman'],as_index=False).ball.count().ball[data2 == True]
plot5 = sns.scatterplot(x= data4, y=data3)
plot5.set(title = 'Runs Scored vs Balls Faced',
         xlabel = 'Ball Faced', ylabel = 'Runs Scored')
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
data1 = deliveries.groupby(['match_id','batsman'],as_index=False).batsman_runs.sum().batsman_runs  
data2 = deliveries.groupby(['match_id','batsman'],as_index=False).ball.count().ball > 14
runs_scored  = data1[data2 == True]
balls_faced = deliveries.groupby(['match_id','batsman'],as_index=False).ball.count().ball[data2 == True]
strike_rate = (runs_scored / balls_faced) * 100

x = pd.DataFrame({'strike_rate':strike_rate,'balls_faced':balls_faced})
minline = x.groupby('balls_faced').min()
maxline = x.groupby('balls_faced').max()


plot6 = sns.scatterplot(x= balls_faced, y=strike_rate)
plot6 = sns.lineplot(x = minline.index , y = minline.strike_rate, color = 'r', label = 'Max Strike Rate')
plot6 = sns.lineplot(x = maxline.index , y = maxline.strike_rate, color = 'g', label = 'Min Strike Rate')
plot6.legend()

plot6.set(title = 'Strike Rate Progression with Balls',
         xlabel = 'Balls Faced', ylabel = 'Strike Rate')
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))

df = pd.merge(matches,deliveries,left_on='id', right_on='match_id')

batsman_runs_total = df.groupby(['batsman'],as_index=False).batsman_runs.sum().batsman_runs
batsman_runs_over_2500 = df.groupby(['batsman'],as_index=False).batsman_runs.sum().batsman_runs >= 2500
balls_faced_total = df.groupby('batsman', as_index=False).ball.count()

balls_faced_total = balls_faced_total [batsman_runs_over_2500 == True]

batsman_runs_total = batsman_runs_total[ batsman_runs_over_2500 == True]
balls_faced_total['runs'] = batsman_runs_total
balls_faced_total['strike_rate'] = (balls_faced_total['runs'] / balls_faced_total['ball']) * 100

balls_faced_total.sort_values(by = 'strike_rate', ascending = False, inplace = True)

plot7 = sns.barplot(x= balls_faced_total['batsman'], y=balls_faced_total['strike_rate'])

plot7.set(title = 'Strike Rate of all batsmen having minimum runs of 2500',
         xlabel = 'Batsman Name', ylabel = 'Strike Rate')

plot7.set_xticklabels(plot7.get_xticklabels(),rotation=90)

plt.show()
```

```python
# 8

fig=plt.figure(figsize=(15,8))

mxa = deliveries.groupby('batsman').agg({'player_dismissed':'count','batsman_runs':sum})
mxa = mxa[mxa['batsman_runs'] > 2499]
mxa[ 'average' ] = mxa['batsman_runs'] / mxa['player_dismissed']
mxa.sort_values(by = 'average', ascending = False, inplace = True)
plot8 = sns.barplot(x= mxa.index, y=mxa.average)
plot8.set(title = 'Average of all batsmen having minimum runs of 2500',
         xlabel = 'Batsman Name', ylabel = 'Average')
plot8.set_xticklabels(plot8.get_xticklabels(),rotation=90)

plt.show()
```

```python
# 9
fig=plt.figure(figsize=(15,8))

df = deliveries.groupby(['match_id','batsman']).agg({'batsman_runs':'sum',
                                                    'ball':'count'})
df = df [ df['batsman_runs'] > 99]
df = df.reset_index()
df = df.groupby('batsman').agg({'batsman_runs':'sum','ball':'sum'})
df['strike_rate'] = (df['batsman_runs'] / df['ball'])*100
df.sort_values(by = 'strike_rate', ascending = False, inplace = True)
df = df.iloc[:10]

plot9 = sns.barplot(x= df.index, y=df.strike_rate, palette ='PuBu_d')
plot9.set(title = 'Strike rate of top 10 batsmen when they score 100',
         xlabel = 'Batsman', ylabel = 'Strike Rate')
plt.show()
```

```python
# 10

fig=plt.figure(figsize=(15,8))

MI = matches [ (matches['city'] == 'Mumbai') & ((matches['team1'] == 'Mumbai Indians') | (matches['team2'] == 'Mumbai Indians'))]
KKR = matches [ (matches['city'] == 'Kolkata') & ((matches['team1'] == 'Kolkata Knight Riders') | (matches['team2'] == 'Kolkata Knight Riders'))]
DC = matches [ (matches['city'] == 'Delhi') & ((matches['team1'] == 'Delhi Capitals') | (matches['team2'] == 'Delhi Capitals'))]
RCB = matches [ (matches['city'] == 'Bangalore') & ((matches['team1'] == 'Royal Challengers Bangalore') | (matches['team2'] == 'Royal Challengers Bangalore'))]
CSK = matches [ (matches['city'] == 'Chennai') & ((matches['team1'] == 'Chennai Super Kings') | (matches['team2'] == 'Chennai Super Kings'))]
RR = matches [ (matches['city'] == 'Jaipur') & ((matches['team1'] == 'Rajasthan Royals') | (matches['team2'] == 'Rajasthan Royals'))]
KXIP = matches [ (matches['city'] == 'Chandigarh') & ((matches['team1'] == 'Kings XI Punjab') | (matches['team2'] == 'Kings XI Punjab'))]
SRH = matches [ (matches['city'] == 'Hyderabad') & ((matches['team1'] == 'Sunrisers Hyderabad') | (matches['team2'] == 'Sunrisers Hyderabad'))]

total_matches = [len(MI), len(KKR), len(DC), len(RCB), len(CSK), len(RR), len(KXIP), len(SRH)]

MI1 = MI[MI['winner'] == 'Mumbai Indians']
KKR1 = KKR[KKR['winner'] == 'Kolkata Knight Riders']
DC1 = DC[DC['winner'] == 'Delhi Capitals']
RCB1 = RCB[RCB['winner'] == 'Royal Challengers Bangalore']
CSK1 = CSK[CSK['winner'] == 'Chennai Super Kings']
RR1 = RR[RR['winner'] == 'Rajasthan Royals']
KXIP1 = KXIP[KXIP['winner'] == 'Kings XI Punjab']
SRH1 = SRH[SRH['winner'] == 'Sunrisers Hyderabad']

win_matches = [len(MI1), len(KKR1), len(DC1), len(RCB1), len(CSK1), len(RR1), len(KXIP1), len(SRH1)]
win_matches

plot10 = sns.barplot(x= ['MI', 'KKR', 'DC', 'RCB', 'CSK', 'RR', 'KXIP', 'SRH'], y =total_matches, color = 'r', label = 'Total Matches')
plot10 = sns.barplot(x= ['MI', 'KKR', 'DC', 'RCB', 'CSK', 'RR', 'KXIP', 'SRH'], y =win_matches, color='b',edgecolor='white', label = 'Wins')


handles, labels = plot10.get_legend_handles_labels()
plot10.legend(handles[::-1], labels[::-1], frameon = False, ncol = 2)

plot10.set(title = 'Home city matches and wins for currently playing 8 teams',
         xlabel = 'Teams', ylabel = 'Wins')

plt.show()
```
